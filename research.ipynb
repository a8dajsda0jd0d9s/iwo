```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
data = pd.read_csv('tweets.csv')


print(data[0:3])
```

```python
#preprocessing
data['text'] = data['text'].apply(lambda x: x.lower())
```

```python
trump_retweets = {'hillary': [], 'other': []}
trump_favorites = {'hillary': [], 'other': []}
hillary_retweets = {'trump': [], 'other': []}
hillary_favorites = {'trump': [], 'other': []}


for index, tweet in data.iterrows():
    if tweet['handle'] == 'realDonaldTrump':
        if 'hillary' in tweet['text'] or 'crooked h' in tweet['text']:
            trump_retweets['hillary'].append(tweet['retweet_count'])
            trump_favorites['hillary'].append(tweet['favorite_count'])
            
        else:
            trump_retweets['other'].append(tweet['retweet_count'])
            trump_favorites['other'].append(tweet['favorite_count'])
            
    else:
        if 'trump' in tweet['text']:
            hillary_retweets['trump'].append(tweet['retweet_count'])
            hillary_favorites['trump'].append(tweet['favorite_count'])
            
        else:
            hillary_retweets['other'].append(tweet['retweet_count'])
            hillary_favorites['other'].append(tweet['favorite_count'])
            
    
    
#print([tweet_stat_sum['trump']['HillaryClinton'], tweet_stat_sum['democrats']['realDonaldTrump']])
```

```python
# print stats
print("Trump:")
print(trump_retweets)
```

# Gemiddelde van retweets

```python
print(np.mean(hillary_retweets['trump']))
print(np.mean(hillary_retweets['other']))
print(np.mean(trump_retweets['hillary']))
print(np.mean(trump_retweets['other']))
```

# Standaard deviatie retweets

```python
print(np.std(hillary_retweets['trump']))
print(np.std(hillary_retweets['other']))
print(np.std(trump_retweets['hillary']))
print(np.std(trump_retweets['other']))
```

```python
print(np.mean(hillary_favorites['trump']))
print(np.mean(hillary_favorites['other']))
print(np.mean(trump_favorites['hillary']))
print(np.mean(trump_favorites['other']))
```
